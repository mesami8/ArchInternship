```python
from google.colab import files
uploaded = files.upload()
audio_path = next(iter(uploaded.keys()))
print("Uploaded:", audio_path)
```

```python
import whisper

# Load small Whisper model (you can change to 'base' or 'medium')
model = whisper.load_model("base")  # or 'small', 'medium', 'large'
result = model.transcribe(audio_path)
transcription = result["text"]

print("🔊 Transcription:")
print(transcription)
```

```python
from unsloth import FastLanguageModel
llm, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/mistral-7b-bnb-4bit",
    max_seq_length=2048,
    dtype=None,  # Let it auto-select
    load_in_4bit=True
)
```

```python
prompt = f"""You are an expert AI. The user just said:

"{transcription}"

Please interpret what the user means and provide a clear, logical response.
"""

inputs = tokenizer(prompt, return_tensors="pt")
inputs = inputs.to(llm.device)

response = llm.generate(**inputs, max_new_tokens=150, temperature=0.7)
decoded_response = tokenizer.decode(response[0], skip_special_tokens=True)

print("🧠 Reasoning Output:")
print(decoded_response)
```
